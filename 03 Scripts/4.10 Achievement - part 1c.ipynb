```python
#Import Libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#Path
path = r'C:\Users\Family Computer\Documents\Data Analytics\Instacart Basket Analysis'
```

```python
#Import orders_products_all.pkl
df_ords_prods_all = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_all_2.pkl'))
```

```python
#Create a scatterplot to see if there is a connection between hour_of_day_ordered and orders_day_of_week
scatterplot_hour_day = sns.scatterplot(x = 'orders_day_of_week', y = 'hour_of_day_ordered', data = df_ords_prods_all, color='green')
```

```python
#Exporting the scatterplot
scatterplot_hour_day.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatterplot_hour_day.png'))
```

```python
#Aggregate sum of orders (prices) by user_id
df_ords_prods_all.groupby('user_id').agg({'prices': ['sum']})
```

```python
df_ords_prods_all.head()
```

```python
#Transform aggregated data to create a new column total_spent
df_ords_prods_all['total_spent'] = df_ords_prods_all.groupby(['user_id'])['prices'].transform(np.sum)
```

```python
df_ords_prods_all.head()
```

```python
# Save a palette to a variable:
palette = sns.color_palette('Greens_r', 10)
```

```python
# Use palplot and pass in the variable:
sns.palplot(palette)
```

```python
diverging_colors = sns.color_palette('Greens', 10)
sns.palplot(diverging_colors)
```

```python
#Key Question #2 total_spent by user_id to hour_of_day_ordered
scatterplot_total_spent_by_hour = sns.scatterplot(x = 'hour_of_day_ordered', y = 'total_spent', data = df_ords_prods_all, color='green')
```

```python
#Exporting the scatterplot
scatterplot_total_spent_by_hour.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatterplot_total_spent_by_hour.png'))
```

```python
#Group by user_id for max_order.
df_ords_prods_all.groupby(['user_id']).agg({'max_order':['mean', 'min', 'max']})
```

```python
#Create scatterplot prices to products
scatterplot_prices_department = sns.scatterplot(x = 'department_name', y = 'prices', data = df_ords_prods_all_merged, color='green')
scatterplot_prices_department.set_xticklabels(scatterplot_prices_department.get_xticklabels(), rotation=45,
                        horizontalalignment='right')
```
