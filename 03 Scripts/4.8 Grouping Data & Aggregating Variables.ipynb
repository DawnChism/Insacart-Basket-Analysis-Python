```python
#Import Libraries
import pandas as pd
import numpy as np
import os
```

```python
#Path
path = r'C:\Users\Family Computer\Documents\Data Analytics\Instacart Basket Analysis'
```

```python
#Import ords_prods_merge_updated dataframe
df_ords_prods_merge_updated = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_updated.pkl'))
```

```python
#2. In this Exercise, you learned how to find the aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe. Now, repeat this process for the entire dataframe.
df_ords_prods_merge_updated.groupby('department_id').agg({'order_number':['mean']})
```

#3. Analyze the result. How do the results for the entire dataframe differ from those of the subset? Include your comments in a markdown cell below the executed code.
##The subset only included results from 8 of the 21 departments. We can now see the actual averages of all 21 departments, whereas when we limited the dataframe to 1,000,000 rows, the averages were also based on those 1,000,000 rows. For example, department 4 had an average of 18.825780 when we only looked at 1,000,000 rows. Now that we are looking at the entire dataframe, we can see that the actual average for department 4 is 17.811403. 

```python
#4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.
df_ords_prods_merge_updated['max_order']=df_ords_prods_merge_updated.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge_updated.head(15)
```

```python
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['max_order'] >40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge_updated.loc[(df_ords_prods_merge_updated['max_order'] <=40) & (df_ords_prods_merge_updated['max_order'] >=10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['max_order'] <=10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge_updated['loyalty_flag'].value_counts (dropna = False)
```

```python
df_ords_prods_merge_updated[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#5. The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.
df_ords_prods_merge_updated.groupby('loyalty_flag').agg({'prices':['mean', 'median', 'min', 'max']})
```

```python
#6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
## If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
df_ords_prods_merge_updated.groupby('product_name')
```

```python
#Marking values above $25.00 as missing - Nan
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['prices'] >100, 'prices']=np.nan
```

```python
df_ords_prods_merge_updated['prices'].max()
```

```python
### Aggregating the data
df_ords_prods_merge_updated.groupby('product_name').agg({'prices':['mean']})
```

```python
### Transform the data
df_ords_prods_merge_updated['average_price']=df_ords_prods_merge_updated.groupby(['user_id'])['prices'].transform (np.mean)
```

```python
df_ords_prods_merge_updated.head(15)
```

```python
### Create a spending flag
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['average_price'] <10, 'spending_flag']= 'Low spender'
```

```python
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['average_price'] >=10, 'spending_flag']= 'High spender'
```

```python
### Value Counts
df_ords_prods_merge_updated['spending_flag'].value_counts (dropna = False)
```

```python
df_ords_prods_merge_updated[['user_id', 'average_price', 'spending_flag']].head(60)
```

```python
#7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
df_ords_prods_merge_updated.groupby('user_id')
```

```python
### Aggregating the data
df_ords_prods_merge_updated.groupby('user_id').agg({'days_since_prior_order':['median']})
```

```python
### Transform the data
df_ords_prods_merge_updated['median_days_since prior_order']=df_ords_prods_merge_updated.groupby(['user_id'])['days_since_prior_order'].transform (np.median)
```

```python
df_ords_prods_merge_updated.head(15)
```

```python
### Create a frequency flag
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['median_days_since prior_order'] >20, 'frequency_flag']= 'Non-frequent customer'
```

```python
df_ords_prods_merge_updated.loc[(df_ords_prods_merge_updated['median_days_since prior_order'] <=20) & (df_ords_prods_merge_updated['median_days_since prior_order'] >10), 'frequency_flag']= 'Regular customer'
```

```python
df_ords_prods_merge_updated.loc[df_ords_prods_merge_updated['median_days_since prior_order'] <=10, 'frequency_flag']= 'Frequent customer'
```

```python
### Value Counts
df_ords_prods_merge_updated['frequency_flag'].value_counts (dropna = False)
```

```python
df_ords_prods_merge_updated[['user_id', 'median_days_since prior_order', 'frequency_flag']].head(60)
```
